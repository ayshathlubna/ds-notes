## Visualizing Data From a Dataframe

```python
col1 = np.linspace(0,10,1000)
col2 = np.sin(col1)

df = pd.DataFrame({'C1' : col1, 'C2' : col2})
df.head(10)
```

```python
df.info()
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot(df.C1, df.C2)
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot(df['C1'], df['C2'])
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot('C1', 'C2', data = df)
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot('C1', 'C2', data = df)
```

```python
plt.style.use('dark_background')
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot('C1', 'C2', data = df)
```

```python
# rc parameters
mpl.rcParamsDefault
```

```python
sns.set(rc = {'axes.facecolor' : '#90d1d0',
              'axes.grid' : True,
              'axes.labelsize' : 14,
              'axes.labelcolor' : 'black',
              'axes.titlesize' : 16,
              'axes.titlecolor' : 'black',
              'xtick.labelsize' : 12,
              'xtick.color' : 'black',
              'ytick.labelsize' : 12,
              'ytick.color' : 'black',
              'figure.facecolor' : 'white'})
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot(df.C1, df.C2)
sns.lineplot(df.C1, df.C2 + 1)

plt.title('Sinewave')
plt.xlabel('X')
plt.ylabel('Sin(x)')
plt.show()
```

```python
auto = pd.read_csv('Automobile_Data.csv')
auto.head(10)
```

```python
auto.info()
```

```python
auto.replace('?',np.nan, inplace = True)
```

```python
auto.duplicated().sum()
```

```python
auto.isnull().sum()
```

```python
auto.price.dtype
```

```python
auto.price = auto.price.apply(float)
auto.horsepower = auto.horsepower.apply(float)
auto['peak-rpm'] = auto['peak-rpm'].apply(float)
```

```python
auto.nunique()
```

```python
print('Fuel types       :', auto['fuel-type'].unique())
print('Body Styles      :', auto['body-style'].unique())
print('Wheel Drives     :', auto['drive-wheels'].unique())
print('Engine Location  :', auto['engine-location'].unique())
print('Engine Types     :', auto['engine-type'].unique())
print('Fuel System      :', auto['fuel-system'].unique())
print('Aspiration       :', auto['aspiration'].unique())
print('Number of Doors  :', auto['num-of-doors'].unique())
print('No. Of Cylinders :', auto['num-of-cylinders'].unique())
```

```python
auto.make.unique()
```

# Distplot

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.distplot(auto.price, hist = False, color = 'red')
```

# Lineplot

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot('engine-size', 'horsepower', data = auto)
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot('horsepower', 'city-mpg', data = auto)
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot('curb-weight', 'city-mpg', data = auto)
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot('curb-weight', 'city-mpg', data = auto, hue = 'fuel-type')
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot('curb-weight', 'city-mpg', data = auto, hue = 'fuel-type', palette = ['red', 'green'])
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot('curb-weight', 'city-mpg', data = auto, style = 'fuel-type')
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.lineplot('curb-weight', 'city-mpg', data = auto, style = 'fuel-type', hue = 'fuel-type')

plt.title('Curb Weight Vs City Mileage')
plt.xlabel('Curb Weight (Kg)')
plt.ylabel('City Mileage (miles/gallon)')
plt.show()
```
