#### 1. Write a program to take principal amount, rate and interest as input and calculate simple interest.

```python
# A program to take principal amount, rate and interest as input and calculate simple interest.

# Accepting the inputs for prinicipal amount, interest rate and number of years

principal = int(input('Enter the prinicipal amount : '))                  
rate = int(input('Enter the rate of interest : '))           
years = int(input('Enter the no.of years for the interest : '))      

# Formula to calculate Simple interest

interest = (principal*rate*years)/100                                  

# Displaying the calculated simple interest
print('\nThe simple interest at the rate of {}% for {} of years = '.format(rate,years), interest)
```

#### 2. Write a program that takes a string as input and prints the number of words in the string.

```python
# A program to take a string as input and prints the number of words in the string.

# Accepting the string
string=input("please enter the string :")

# Spliting the string into list of words by a white space ' '
listofwords=string.split(' ')

# Printing the list of words extracted from the string
print("\nThe list of words from the string : ",listofwords)

# Counting and printing the number of words in the string
print("\nNo.of words present in the string : ",len(listofwords))
```

#### 4. Write a program to take the marks of 3 students in five subjects each as a list (one list for each input) as input and print the average score of each student..

```python
# a program to take the marks of 3 students in five subjects each as a list as input and print the average score of each student

# Accepting the marks of 3 students in 5 subjects
student1 = input("Enter the marks of student 1 : ")
student2 = input("Enter the marks of student 2 : ")
student3 = input("Enter the marks of student 3 : ")

# Making a list of marks of each student
student1=student1.split(' ')
student2=student2.split(' ')
student3=student3.split(' ')

# Converting the marks in the list to integer format
student1=list(map(int,student1))
student2=list(map(int,student2))
student3=list(map(int,student3))

# Printing the list of marks of each student
print("\nList of Marks of student1 : ",student1)
print("List of Marks of student2 : ",student2)
print("List of Marks of student3 : ",student3)

# calculating the average mark of each student and printing their result
avg_student1=sum(student1)/5
avg_student2=sum(student2)/5
avg_student3=sum(student3)/5

print("\nAverage mark of student1 : ",avg_student1)
print("Average mark of student2 : ",avg_student2)
print("Average mark of student3 : ",avg_student3)
 
```

#### 5. Write a program to take a list as input containing results of 10 coin tosses as 'H' or 'T'. Jon gets a point for each 'T' and Tom gets a point for each 'H'. print the points recieved by both.

```python
# A program to calculate the points scored by 2 people in a coin toss game

# Accepting and printing the result of the game

result=list(input("Enter the result of coin tosses : "))
print("Result of the game : ",result)

# counting the no.of points scored by each participant

tom_points = result.count('H')
jon_points = result.count('T')

# Printing the no.of points scored by each participant

print("\nNo.of points Tom scored : ",tom_points)
print("No.of points Jon scored : ",jon_points)
```

#### 6. Take two numbers as input and print the result after swapping them. (Without using multiple assignment and any extra variable).

```python
# A program to take two numbers as input and print the result after swapping them

# Entering the value of 2 numbers
num1 = int(input('Enter First Number  : '))
num2 = int(input('Enter Second Number : '))

# printing the numbers before swapping
print('\nBefore Swapping : ')
print('Num1 = ',num1)
print('Num2 = ',num2)

# Swapping the two numbers by adding the 2 numbers first and then substracting the numbers from the sum
num1 = num1 + num2
num2 = num1 - num2
num1 = num1 - num2

# Printing the values after swapping

print('\nAfter Swapping : ')
print('Num1 = ',num1)
print('Num2 = ',num2)
```

#### 7. Take a list as input in a single line containing 10 numbers as marks out of 100. The even positions in the list are marks of Tom and the odd positions are the marks of Jon. Calculate the average marks of both.

```python
# A program to find the marks scored by 2 students from a list of marks and find their average marks

# accepting the marks scored by 2 students

marks=input("Enter the marks : ")

# storing the marks in the form of list

marks=marks.split(' ')
print("\nThe marks scored by the students : ",marks)

# Finding the marks scored by the students from the list based on their index number

tom_marks=marks[0::2]
jon_marks=marks[1::2]

# Calculating the Average marks and Printing the marks and their average

tom_marks=list(map(int,tom_marks))
tom_avgmarks=sum(tom_marks)/5
print("\nMarks scored by Tom        : ",tom_marks)
print("Average mark scored by Tom : ",tom_avgmarks)

jon_marks=list(map(int,jon_marks))
jon_avgmarks=sum(jon_marks)/5
print("\nMarks scored by Jon        : ",jon_marks) 
print("Average mark scored by Jon : ",jon_avgmarks)




```

#### 8. Take a string and a character as input. Print the result having the character between each word of the given string.

```python
''' A program to take a string and a character as input and Print the result having the 
    character between each word of the given string.'''


# Accepting the string and the character
a=input("Enter the string    : ")
b=input("Enter the character : ")

# Seperating the words of the string with white space and printing the result
c=a.split(' ')
print("\nThe words in the string are : ",c)

# Joining the string back with the entered character between each word
d=b.join(c)
print("\nThe string with {} as character between each word  : ".format(b) , d)

```
