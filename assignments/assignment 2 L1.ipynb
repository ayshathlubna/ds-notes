### 1.

Go through the string below and if the length of a word is even print "even!"

string = 'I know how to print every word in this sentence that has an even number of letters'

```python
# A program to find  if the length of a word is even and print "even!" if it is

string = 'I know how to print every word in this sentence that has an even number of letters'.split(' ')

# printing the list of words in the string

print("The list of words in the string : ", string)

print("\n")

#calculating the length of each word in the string and print if the length is even

for i in string:
    if len(i) %2 == 0:
        print( " '{}' have Even no.of letters in it".format(str(i)))

```

### 2. 

Create a function ``lets_find_88`` Given a list of ints, return True if the array contains a 8 next to a 8 somewhere.
- lets_find_88([1, 8, 8]) → True
- lets_find_88([1, 8, 1, 8]) → False
- lets_find_88([8, 1, 8]) → False

```python
# a program to create a function and return True if the array contains a 8 next to a 8 somewhere

# creating 3 lists
list1 = [1,8,8]
list2 = [1,8,1,8]
list3 = [8,1,8]

# checking whether the array has a 8 next to a 8 using function
def lets_find_88(x):
    for i in range(0,len(x)):
        if x[i]==8:
            if x[i+1]==8:
                return True
            else:
                return False
        elif x[i]!=8 and x[i+1]!=8:
                return False
            
# printing the output by passing the arguments to the function
            
print("The list " + str(list1) + " has an 8 next to a 8 in the list : ",lets_find_88(list1))
print("The list " + str(list2) + " has an 8 next to a 8 in the list :",lets_find_88(list2))
print("The list " + str(list3) + " has an 8 next to a 8 in the list :",lets_find_88(list3))
```

### 3.

 Calculate and display the sum of all the numbers divisible by 7 between 18 and 534.
 
 Hint: One way to do this is to loop over a 'range' using 'for' loop and use an 'if' statement inside it.

```python
# Calculate and display the sum of all the numbers divisible by 7 between 18 and 534.

# initialising the variable sum as zero
sum=0

# printing all the numbers divisible by 7 between 18 and 534 and finding its sum

print('The numbers divisible by 7 between 18 and 534 are : \n ')

for i in range(18,534,1):
   
    if i % 7 == 0:
        sum=sum+i
        print(i,end = ' ')

# printing the sum of all the numbers that are divisible by 7 between 18 and 534
print("\n\nsum of all the above numbers : ",sum)
        
```

Write a function to return True if the first and last number of a given list is same. If numbers are different then return False.

- [11, 13, 17, 19, 11] --> True
- [24, 36, 12, 31, 42] --> False

```python
# a program to return True if the first and last number of a given list is same

list1=[11,13,17,19,11]
list2=[24,36,12,31,42]

# creating a function to check whether the first and last element in the list is same or not
def check(x):
    
    if x[0]==x[-1]:
        return "True"
    else:
        return "False"
    
# printing the output by passing the arguments to the function
print("The first and last element in "+ str(list1) + " is same: ",check(list1))
print("The first and last element in "+ str(list2) + " is same: ",check(list2))
```
