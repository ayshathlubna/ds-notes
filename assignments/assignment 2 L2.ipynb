#### 1. Take a list of numbers as input. Print whether the average is even or odd.

```python
# A program to take a list of numbers as input and print whether the average is even or odd.

# Accepting the numbers into list

list1=input("Enter the numbers into the list : ").split(" ")
n=len(list1)

list1=list(map(int,list1))

print("The entered list is    : ", list1)

# Finding the average and printing the result

average = sum(list1)/int(n)
print("\nAverage of the numbers : ", average)

#checking whether the average is even or odd

if average% 2 == 0 :
    print("The average of the numbers is even ")
else:
    print("The average of the numbers is odd ")
    
    
```

#### 3. Take four numbers as input in a single line. Find out the largest number.

```python
# A program to take four numbers as input in a single line and find out the largest number.

# accepting the numbers and converting to list of integers
num=input("Enter four numbers : ").split(" ")

a,b,c,d=list(map(int,num))

# finding out the largest among the four numbers using if statement

if a>=b and a>=c and a>=d:
    print(a," is the largest number")
    
elif b>=c and b>=d:
    print(b," is the largest number")
    
elif c>=d:
    print(c," is the largest number")
    
else:
    print(d," is the largest number")
    
```

#### 5. Create a dictionary having three items. Key of each item is the name of a student, value associated with each key is a list of marks in five subjects. Given this dictionary print the top and lowest scorers.

```python
# A program to find the top scorer and lowest scorer among 3 students

student_marks = {"Shanza" : [45,36,47,38,41], "Ahyaan" : [47,49,45,42,46], "Izyaan" : [39,41,43,40,45]}

# Extracting the marks of each students from dictionary

name_student1,name_student2,name_student3=list(student_marks.keys())
student1,student2,student3=list(student_marks.values())

print("Marks of " + name_student1 + ": ",student1)
print("Marks of " + name_student2 + ": ",student2)
print("Marks of " + name_student3 + ": ",student3)

#finding the total marks of each student and displaying their result
total_student1=sum(student1)
total_student2=sum(student2)
total_student3=sum(student3)

print("\nTotal Marks of " + name_student1 + ": " ,total_student1)
print("Total Marks of " + name_student2 + ": " ,total_student2)
print("Total Marks of " + name_student3 + ": " ,total_student3)

print("\n")

# finding the top scorer and lowest scorer using if condition

if total_student1>=total_student2 and total_student1>=total_student3:
    print(name_student1,"is the top scorer")
elif total_student2>=total_student3:
    print(name_student2,"is the top scorer")
else:
    print(name_student3,"is the top scorer")

if total_student1<=total_student2 and total_student1<=total_student3:
    print(name_student1,"is the lowest scorer")
elif total_student2<=total_student3:
    print(name_student2,"is the lowest scorer")
else:
    print(name_student3,"is the lowest scorer")
```

#### 7. Take a list as input in a single line containing 15 numbers as marks out of 100. The first five numbers in the list are marks of Tom, the middle 5 numbers are marks of Bob and the last 5 are the marks of Jon. Find out the top and lowest scorer.

```python
# a program to find the top and lowest scorer among 3 students in 5 subjects

marks=input("Enter the marks : ").split(" ")
marks=list(map(int,marks))
print("List of marks scored by 3 students : ",marks)


# he first five numbers in the list are marks of Tom, the middle 5 numbers are marks of Bob and the last 5 are the marks of Jon
tom_marks=marks[0:5]
bob_marks=marks[5:10]
jon_marks=marks[10:15]

print("\nMarks of Tom : ",tom_marks)
print("Marks of Bob : ",bob_marks)
print("Marks of Jon : ",jon_marks)

#finding the total marks of each student

tom_total=sum(tom_marks)
print("\nTotal of Tom's marks : ",tom_total)
bob_total=sum(bob_marks)
print("Total of Bob's marks : ",bob_total)
jon_total=sum(jon_marks)
print("Total of Jon's marks : ",jon_total)

print("\n")

#checking which student has scored an overall highest and lowest in 5 subjects

if tom_total>=bob_total and tom_total>=jon_total:
    print("Tom is the top scorer")
elif bob_total>=jon_total:
    print("Bob is the top scorer")
else:
    print("Jon is the top scorer")

if tom_total<=bob_total and tom_total<=jon_total:
    print("Tom is the lowest scorer")
elif bob_total<=jon_total:
    print("Bob is the lowest scorer")
else:
    print("Jon is the lowest scorer")
```
