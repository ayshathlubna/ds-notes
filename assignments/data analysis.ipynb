```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Haris', 'Sam', 'Rick', 'Maggy', 'Daryl', 'Robita', 'Negan', np.nan],
    'Age': [28, np.nan, 35, 30, 40, np.nan, 45, 38],
    'Salary': [50000, 60000, np.nan, 55000, 70000, 80000, np.nan, 90000],
    'Department': ['HR', 'IT', 'Sales', np.nan, 'HR', 'Marketing', 'IT', 'Sales'],
    'Gender': ['Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', np.nan]
}

# Create a DataFrame
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
# Mean Imputation for 'Age' column
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

```python
# Median Imputation for 'Salary' column
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df
```

```python
# Mode Imputation for 'Department' column
df['Department'].fillna(df['Department'].mode()[0], inplace=True)
df
```

```python
# Creating a new 'Unknown' category for missing values in 'Gender'
df['Gender'].fillna('Unknown', inplace=True)
df
```

```python
df
```

```python
from sklearn.impute import KNNImputer

# Using KNN Imputer to fill missing values
imputer = KNNImputer(n_neighbors=2)
df_imputed = pd.DataFrame(imputer.fit_transform(df[['Age', 'Salary']]), columns=['Age', 'Salary'])

df[['Age','Salary']] = df_imputed
df
```

```python
df
```

```python
# Dropping rows with missing values in the 'Name' column
df_dropped = df.dropna(subset=['Name'])
df_dropped
```

```python
data = {
    'Name': ['Haris', 'Sam', 'Rick', 'Maggy', 'Daryl', 'Robita', 'Negan', np.nan],
    'Age': [28, 29, 35, 30, 40, 29, 45, 38],
    'Salary': [50000, 60000, np.nan, 55000, 70000, np.nan, 85000, 90000],
    'Department': ['HR', 'IT', 'Sales', np.nan, 'HR', 'Marketing', 'IT', 'Sales'],
    'Gender': ['Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male', np.nan]
}

# Create a DataFrame
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

## Preprocessing

```python
# Convert categorical variables using one-hot encoding
df_encoded = pd.get_dummies(df[['Department', 'Gender']], drop_first=True)

# Add the one-hot encoded columns back to the original DataFrame
df = pd.concat([df[['Age', 'Salary']], df_encoded], axis=1)

# Display the updated DataFrame
df
```

## Split Data into Training and Prediction Sets

```python
# Split the dataset into training and prediction sets for 'Salary'
train_data = df[df['Salary'].notna()]  # Data where Salary is not missing
predict_data = df[df['Salary'].isna()]  # Data where Salary is missing

# Separate features and target variable
X_train = train_data.drop('Salary', axis=1)
y_train = train_data['Salary']

X_predict = predict_data.drop('Salary', axis=1)
```

## Train a Regression Model

```python
from sklearn.linear_model import LinearRegression

# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict the missing Salary values
predicted_salaries = model.predict(X_predict)
predicted_salaries
```

```python
# Assign the predicted salaries back to the DataFrame
df.loc[df['Salary'].isna(), 'Salary'] = predicted_salaries

# Display the DataFrame with filled Salary values
df
```

## Evaluating the result
