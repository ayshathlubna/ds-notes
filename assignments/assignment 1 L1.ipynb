### 1.

Write a Python program to get a list, sorted in increasing order by the last element in each tuple from a given list of non-empty tuples.

Sample List : [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
Expected Result : [(2, 1), (1, 2), (2, 3), (4, 4), (2, 5)]

```python
# A Python program to get a list, sorted in increasing order by the last element in each tuple from a given list of non-empty tuples.

# creating a list of non-empty tuples and printing it

lists= [(3, 8), (2, 7), (6, 4), (2, 5), (1, 3)] 
print("List before sorting : ", lists)

# sorting the list by the last element in each tuple in increaing order

lists.sort(key=lambda y:y[1])
print("\nList after sorting by the last element in each tuple : ",lists)
```

### 3.

Write a program to accept a string from the user and display characters that are present at an even index number.

For example, str = "judgemental" so you should display ‘j’, ‘d’, ‘e’, ‘e’, ‘t’, ‘l’.

```python
# A program to accept a string from the user and display characters that are present at an even index number

# Accepting string from the user
str=input("Please enter the string : ") 

# printing the entered string
print("\nThe entered string is   :",str)

# Extracting and displaying the characters present at the even index number from the string

str2=tuple(str[0::2])
print("\nThe characters present at the even index number is : ",str2) 

```

### 4. 

Convert the Float Value of '3.8' into Integer, Complex, Binary and Hexidecimal format

```python
# A program to Convert the Float Value of '3.8' into Integer, Complex, Binary and Hexidecimal format

# converting the float number to integer and complex format

x=3.8
print("Integer value of 3.8 : ",int(x)) #converting to integer value
print("Complex value of 3.8 : ",complex(x)) #converting to complex value

#converting the float to integer and then to binary and hexadecimal format

y=int(x)
print("Binary value of 3.8  : ",bin(y)) #converting the integer value to binary
print("Hexadecimal value of 3.8 : ",hex(y)) #converting the integer value to hexadecimal
```
