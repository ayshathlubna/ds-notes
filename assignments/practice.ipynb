##### Create a function which takes a 2-D list of shape m * n where m is number of elements in the original list and where each element is a list having n integers (A list of m lists having n numbers each). The function also a takes a number x as input and returns a list of all consecutive pairs having sum eaqqual to x.


```python
def consecutive_pairs(list1, num):
    pairs = []
    for i in range(len(list1)):
        for j in range(len(list1[i])-1):
            if list1[i][j]+list1[i][j+1]==num:
                t=(list1[i][j],list1[i][j+1])
                pairs.append(t)

    return pairs
m=int(input("Enter the no.of rows    : "))
n=int(input("Enter the no.of columns : "))

matrix=[]
for i in range(m):
    row=[]
    
    for j in range(n):
        num=int(input("enter the elements into the matrix :"))
        row.append(num)
    matrix.append(row)
    
print("matrix : ")

for i in range(m):
    for j in range(n):
        print(matrix[i][j],end=" ")
    print()
    
    
number=int(input("Enter the number : "))
print("The list of consecutive pairs whose sum is {} ".format(number),consecutive_pairs(matrix,number))
```
