## classes and objects

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def display_info(self):
        return f"Student: {self.name}, Age: {self.age}"
```

```python
class GraduateStudent(Student):
    def __init__(self, name, age, degree):
        super().__init__(name, age)                        # Inherits name and age
        self.degree = degree

    def display_info(self):
        return f"{super().display_info()}, Degree: {self.degree}"
```

```python
grad_student = GraduateStudent("Emma", 24, "Master's")
print(grad_student.display_info())
```

## Building a simple student management system

```python
class Student:
    def __init__(self, name, age, grades):
        # Constructor to initialize the attributes for each Student object
        self.name = name                          # Attribute to store the student's name
        self.age = age                            # Attribute to store the student's age
        self.grades = grades                      # Attribute to store the student's list of grades
    
    def calculate_average(self):
        # Method to calculate the average grade of the student
        return sum(self.grades) / len(self.grades)

class School:
    def __init__(self, school_name):
        # Constructor to initialize the School object
        self.school_name = school_name           # Attribute to store the school's name
        self.students = []                       # Attribute to store a list of students in the school
    
    def add_student(self, student):
        # Method to add a student to the school's student list
        self.students.append(student)
    
    def show_students(self):
        # Method to print the names and average grades of all students in the school
        for student in self.students:
            print(f"{student.name} has an average grade of {student.calculate_average()}")
```

```python
# Creating an instance of the School class with the name "XYZ University"
school = School("XYZ University")

# Creating two instances of the Student class with names, ages, and lists of grades
student1 = Student("Henry", 20, [85, 90, 78])  
student2 = Student("Lidia", 21, [88, 92, 81])  

# Adding the student1 and student2 objects to the school's list of students
school.add_student(student1)
school.add_student(student2)

# Displaying the students' names along with their average grades
school.show_students()
```

## Solving pascals triangle


### 1. 

Write a Python function that prints out the first n rows of Pascal's triangle.

- Input : 5
- Output: 
          1
         1 1
        1 2 1
       1 3 3 1
      1 4 6 4 1

```python
def print_pascals_triangle(n):                                      # creating a function
    for i in range(n):
        # Print leading spaces
        print(' ' * (n - i - 1), end='')

        # Calculate and print each element in the row
        num = 1
        for j in range(i + 1):
            print(num, end=' ')
            num = num * (i - j) // (j + 1)

        # Move to the next line
        print()

# Example usage:
print_pascals_triangle(5)
```

## 2.
Write a program to print prime numbers from 1 to 100

```python
def prime():
    for i in range(2,100):
        for j in range(2,int(i**(1/2))+1):
            if i%j==0:
                break
        else:
            print(i,end=" ")
prime()
```

## 5.
Tom and Bob are playing a game of dice. Each time a multiple of two comes up Tom gets a point and if a multiple of 3 comes up bob gets a point. If a multiple of both comes up then a point is deducted from both. Write a program to take a list as input consisting of 10 outcomes of the dice throw. Print the scores of both Tom and bob and the winner.

```python
score = list(map(int,input("Enter the scores : ").split()))

if len(score) != 10:
    print("Please enter exactly 10 numbers.")
else:
    tom_score= 0
    bob_score = 0

    for i in score: 
        if i%2==0 and i%3==0:
            tom_score-=1
            bob_score-=1
        elif i%2==0:
            tom_score+=1
        elif i%3==0:
            bob_score+=1

    print("Tom's score : ",tom_score)
    print("Bob's score : ",bob_score)

    if bob_score>tom_score:
        print(f"Bob is the winner with points {bob_score}")
    elif tom_score>bob_score:
        print(f"Tom is the winner with points {tom_score}")
    else:
        print("Both Tom and Bob have same score")
```

## Build a Simple Calculator using Exception Handling (try, except) that solves 4 arithmetic operations like Addition, Subtraction, Multiplication and Division.
        

```python
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    
    def addition(self):
        return self.a + self.b
    
    def subtraction(self):
        return self.a - self.b
    
    def multiplication(self):
        return self.a * self.b

    def division(self):
        try:
            return self.a / self.b
        except ZeroDivisionError:
            return "Division by '0' is not possible"

# Input numbers from the user
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

# Creating an object of the Calculator class
obj = Calculator(a, b)

# Loop to perform operations
while True:
    op = input("\nEnter the operator (+, -, *, /) or '0' to exit: ")
    if op == '+':
        print(f"Sum of {a} and {b}: {obj.addition()}") 
    elif op == '-':
        print(f"Difference of {a} and {b}: {obj.subtraction()}") 
    elif op == '*':
        print(f"Product of {a} and {b}: {obj.multiplication()}") 
    elif op == '/':
        print(f"Quotient of {a} and {b}: {obj.division()}") 
    elif op == '0':
        print("Exit")
        break
    else:
        print("Please enter a valid operator (+, -, *, /)")
```
