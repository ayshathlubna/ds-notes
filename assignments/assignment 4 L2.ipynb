#### 1. Create a function that takes a list of numbers as input and returns the median.

```python
# a program to create a function that takes a list of numbers as input and returns the median.

def median(num): # defining a finction
    num.sort()   # sorting the list to find the median
    
    print("\nThe sorted list : ", num)
    n = len(num)
    
    if n%2 != 0:
        med = num[(int((n+1)/2)-1)] # if length of list is even using the mentioned formula to find median
        return med
    else:
        med = (num[int(n/2)-1]+num[int((n/2)+1)-1])/2 # if length of list is odd using this method to find median
        return med

list1 = [40,50,10,20,30,60]
list2 = [50,10,30,20,40]

print("The median of the list : ",median(list1))
print("The median of the list : ",median(list2))
```

#### 3. Create a function that takes a list of numbers as input and returns the standard deviation.

```python
# a program to create a function that takes a list of numbers as input and returns the standard deviation.

def standard_deviation(num):
    
    n = len(num)
    m = sum(num)/len(num) # finding the mean
    s = 0
    
    for i in range(0,len(num)):
        x = (num[i]-m)**2
        s = s+x  
    sd = (s/n)**(1/2)
    return sd

list1 = [40,50,10,20,30,60,90,70,80]

print("The standard deviation of the list : ",standard_deviation(list1))
```

#### 4. Create a function that takes a number as input and return True if the number is prime , False if not.

```python
def prime(num):

    if num < 2:      # since the prime number starts from 2 
        return False
    else:
        for i in range(2, int(num ** (1/2)) + 1):
            if num % i == 0:    # if the reminder of the number divided by any number is zero then it is not prime
                return False
        else:
            return True
        
n = int(input('Enter the number : '))
print("The number {} is prime : ".format(n) ,prime(n))
```

#### 5. Create a function that takes a list of numbers as input and returns the variance.

```python
# a program to create a function that takes a list of numbers as input and returns the variance.

def variance(num): # defining a function named variance
    
    n = len(num)
    m = sum(num)/len(num) #finding the mean
    s = 0
    
    for i in range(0,len(num)): #using for loop to iterate through all the elements of the list
        x = (num[i]-m)**2
        s = s+x  
    var = s/n
    return var

list1 = [40,50,10,20,30,60,90,70,80]

print("The variance of the list : ", variance(list1))
```

#### 6. Create a function that takes a number as input and returns the factorial.

```python
# a program to find the factorial of a number

def factorial(num):
    f = 1                 # initialising first number as 1
    
    for i in range(1,num+1):
        f = f*i          # will mutliply all the numbers upto the entered number
    return f

n=int(input("Enter the number : "))

print("The factorial of the number {} : ".format(n),factorial(n))
```

#### 7. Create a function that takes two lists of same length as input and returns a list consisting of larger element at each position after comparing the corresponding elements in the given lists.

```python
# a program to return a list having the largest number at each positions comparing the two list

def larger_list(num1,num2):  # defining a function
    num3 = []                # creating an empty list to store the elements
    
    for i in range(0,len(num1)):
        if num1[i] >= num2[i]:  # comparing the elements of both the lists based on their index and storing the latgest one in 3rd list
            num3.append(num1[i])
        else:
            num3.append(num2[i])
    return num3
            
list1=[8,3,6,1,2,5,9,4,7]
list2=[6,9,8,1,3,7,4,2,5]

print("The list which has the largest elements at each position of the two lists : ",larger_list(list1,list2))
```

#### 11. Create a function which takes a 2-D list of shape m * n where m is number of elements in the original list and where each element is a list having n integers (A list of m lists having n numbers each). The function also takes a number x as input and returns True is x is in the given 2D list.

```python
# a program to check whether a number is present in the 2D list

def num_present(lists,x):            # defining a function named num_present
    
    for i in lists:                  # creating i as element of the matrix
        if x in i:                   # checking if the number matches with the element of the matrix 
            return True
    return False       


# accepting the numnber of rows and columns
m = int(input("Enter the no.of rows    : "))
n = int(input("Enter the no.of columns : "))


matrix = []     # creating an empty matrix
for i in range(m):
    row = []    # creating an empty row
    
    for j in range(n):
        num = int(input("enter the elements into the matrix :"))  # accepting the elements into the row
        row.append(num)          # adding the elements to the row
    matrix.append(row)           # adding the row to the matrix
    
print("matrix : ")

for i in range(m):
    for j in range(n):
        print(matrix[i][j],end=" ")   # printing the matrix
    print()
    
element = int(input("Enter the number to check if present in the list : "))
print("The number is present : ",num_present(matrix,element))
```

```python
# a program to find the number of words in a file

def count_words(file1):               # defining a function           
    string = open(file1).read()       # reading the contents of file and storing in variable string
    words = string.split()            # splitting the contents into list of words 
    num_words = len(words)            # counting the length of the list
    return num_words

print("Number of words in the file : ",count_words("program.txt"))
```

#### 16. Create a function which takes a txt filename as input and returns the number of line in the file.

```python
# a program to find the number of lines in a file

def count_lines(files):               # defining a function
    string=open(files).read()         # reading the contents of the file 
    lines=string.split("\n")          # splitting the contents of the file by line and storing in a list
    num_lines=len(lines)              # counting the length of the list
    return num_lines


print("Number of lines in the file : ",count_lines("program.txt"))
```

#### 17. Create a function that takes two txt filenames and returns a file after merging the cintents of two given files. (Print the merged content and then return the filename, content of original files should be unchanged).

```python
# a program to merge the contents of 2 files into a third file

def merge(file1 ,file2 ,file3):  # defining a function
    
    file_1=open(file1)
    file1_data=file_1.read()   # reading the file file1 into variable file1_data
    
    file_2=open(file2)
    file2_data=file_2.read()   # reading the file file2 into variable file2_data
    
    file_3=open("file3.txt","w")  # creating a new file in write mode
    file_3.write(file1_data)      # writing the contents of file1 to file3
    file_3.write(file2_data)      # writing the contents of file2 to file3
    
    file_1.close()
    file_2.close()
    
    return file3
   
print("Printing the contents of file 1:\n",open("file1.txt").read())    
print("Printing the contents of file 2:\n",open("file2.txt").read())    
print("Printing the contents of merged file :\n",open("file3.txt").read())    

print("Merged File name : ",merge("file1.txt","file2.txt","file3.txt")) # calling the function which returns the name of the file

```

#### 18. Write a program that takes a string as input and returns a list of all words having even length in that string (Using Map/Filter and Lambda expression).

```python
# a program to return a list of all words in a list having even length using filter and lambda expression

words = input("Enter the string : ").split()
even_words = list(filter(lambda x: len(x) % 2 == 0, words))
print(even_words)
```

```python
# a program to return a list of all words in a list having even length using map, filter and lambda expression

words = input("Enter the string : ").split()
even_words = list(map(lambda x:x,filter(lambda x: len(x) % 2 == 0, words)))
print(even_words)

```
