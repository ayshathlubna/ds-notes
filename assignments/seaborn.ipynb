## Importing the library

```python
import seaborn as sns
```

## Loading the dataset

```python
df = sns.load_dataset("tips")
```

```python
df.head()                                      # seeing the first 5 rows from the dataset
```

```python
df.info()                                      # viewing a summary of the dataset
```

## Line plot

```python
sns.lineplot(x='size', y='total_bill', data=df)
```

## Bar Plot

```python
sns.barplot(x='day', y='tip', data=df)
```

## Count Plot

```python
sns.countplot(x='day', data=df)
```

## Scatter Plot

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
```

## Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

## Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=df)
```

## Pair Plot

```python
sns.pairplot(df)
```

## Heatmap

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

## Strip Plot

```python
sns.stripplot(x='day', y='total_bill', data=df)
```

## Swarm Plot

```python
sns.swarmplot(x='day', y='total_bill', data=df)
```

## Point Plot

```python
sns.pointplot(x='day', y='tip', data=df)
```

## Joint Plot

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

## Facet Grid

```python
g = sns.FacetGrid(df, col="day", row="time")
g.map(sns.scatterplot, "total_bill", "tip")
```
