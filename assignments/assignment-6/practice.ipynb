```python
def insert_data():
    print('\nEnter the details of the employee to be addaed : ')
    Empid = int(input('Enter the employee id : '))
    Fname = input('Enter the first name : ')
    Lname = input('Enter the last name  : ')
    Age = int(input('Enter the age         : '))
    Gender = input('Enter the gender    : ')
    Salary = input('Enter the salary    : ')
    City = input('Enter the city      : ')
    Deptid=input("Enter the department id : ")
    
    testdb_connector = mysql.connector.connect(
                        host = 'localhost',
                        user = 'root',
                        password = 'mysqlpwd123',
                        database ='assignment2')

    mycursor = testdb_connector.cursor()
    
    query='''INSERT INTO employee(Empid,Fname,Lname,Age,Gender,Salary,City,Deptid)
         VALUES (%s,%s,%s,%s,%s,%s,%s,%s)'''
    val=(Empid,Fname,Lname,Age,Gender,Salary,City,Deptid)
    mycursor.executemany(query, val)
    
    testdb_connector.commit()
    testdb_connector.close()
    print('\n Row inserted Successfully')
```

```python
# updating the salary of employee with empid 101

def update_data():
    print('\nEnter the update details : ')
    print()
    column = input('Enter the column : ')
    value  = input('Enter the value : ')
    empid  = int(input('Enter the employee id : '))
    
    testdb_connector = mysql.connector.connect(
                        host = 'localhost',
                        user = 'root',
                        password = 'mysqlpwd123',
                        database ='assignment2')

    mycursor = testdb_connector.cursor()
    
    if column == 'Salary':
        query = "UPDATE employee SET {} = {} WHERE empid = {}".format(column, int(value), empid)
    else:
        query = "UPDATE employee SET {} = '{}' WHERE empid = {}".format(column, value, empid)
        
    mycursor.execute(query)
    testdb_connector.commit()
    testdb_connector.close()
    print('\nUpdate Successfull')
```
