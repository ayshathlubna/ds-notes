### Handling Databases and CRUD operations using Python

###### Create a table in the employee database and insert the employee related data of 6 employees to the employee table. Get all the employee related data and update an employee record into the table. Delete the last employee record of the table.


```python
# importing the sql connector
import mysql.connector
```

```python
# connecting to the host
mydb= mysql.connector.connect(
                            host = 'localhost',
                            user='root',
                            password = 'mysqlpwd123')
```

```python
mycursor=mydb.cursor()
type(mycursor)
```

```python
# creating the database employee
mycursor.execute('CREATE DATABASE employee')
```

```python
# connecting with the employee database
employee_connector= mysql.connector.connect(
                            host = 'localhost',
                            user='root',
                            password = 'mysqlpwd123',
                            database ='employee')
mycursor=employee_connector.cursor()
```

```python
# creating the table employees in employee database
query ='''CREATE TABLE employees(
            empid INT PRIMARY KEY,
            empname VARCHAR(100),
            age INT,
            department VARCHAR(100),
            salary INT)'''
mycursor.execute(query)
```

```python
# showing the tables in the employee database
mycursor.execute('SHOW TABLES')
for table in mycursor:   
    print(table)
```

```python
# inserting the datas to the table
query='''INSERT INTO employees(empid,empname,age,department,salary)
         VALUES (%s,%s,%s,%s,%s)'''

val=[(101,'Ayshath', 35,'HR',100000),
     (102,'Musthafa', 58,'Accounts',110000),
     (103,'Laila', 46,'Finance',90000),
     (104,'Sajid', 39,'Marketing',120000),
     (105,'Mufeeda', 32,'Administration',130000),
     (106,'Abdulla', 29,'Sales',80000)]

mycursor.executemany(query,val)
```

```python
# viewing the table employees from the employee database
query ='SELECT * FROM employees'
mycursor.execute(query)
for row in mycursor:
    print(row)
```

```python
# Updating the salary of the employees whose employee id is 102
query = 'UPDATE employees SET salary=150000 WHERE empid=102'
mycursor.execute(query)
```

```python
# viewing the updated table employees.
query ='SELECT * FROM employees'
mycursor.execute(query)
for row in mycursor:
    print(row)
```

```python
# Deleting the information of the last employee in the table
query= 'DELETE FROM employees ORDER BY empid DESC LIMIT 1'
mycursor.execute(query)
```

```python
# viewing the updated table employees.
query ='SELECT * FROM employees'
mycursor.execute(query)
for row in mycursor:
    print(row)
```
