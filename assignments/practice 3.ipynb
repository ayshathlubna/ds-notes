## Calculate the average temperature from a dataset

```python
import numpy as np

# Temperature data for a week (in Celsius)
temperatures = np.array([22.5, 21.0, 23.5, 20.0, 19.5, 24.0, 22.0])

# Calculate average temperature
average_temp = np.mean(temperatures)
print(f"Average Temperature : {average_temp}°C")

# Find the maximum and minimum temperatures
max_temp = np.max(temperatures)
min_temp = np.min(temperatures)
print(f"Highest Temperature : {max_temp}°C")
print(f"Lowest Temperature  : {min_temp}°C")

# Temperature difference
temp_diff = np.ptp(temperatures)
print(f"Temperature Difference  : {temp_diff}°C")
```

### Numpy vs Python lists

```python
import time

# Python List
list1 = range(1000000)
start = time.time()
result = [x * 2 for x in list1]
end = time.time()
print(f"Python list time: {end - start} seconds")

# NumPy Array
arr = np.arange(1000000)
start = time.time()
result = arr * 2
end = time.time()
print(f"NumPy array time: {end - start} seconds")
```

## Analyzing Sales Data

```python
import pandas as pd

# Load sales data
sales_data = {'Product_ID': [101, 102, 103, 104, 105, 106],
              'Units_Sold': [200, 150, 50, None, 500, 450],
              'Sale_Amount': [5000, 4000, 1500, 2000, None, 12000]}

sales_df = pd.DataFrame(sales_data)

# Load product information data
product_data = {'Product_ID': [101, 102, 103, 104, 105, 106],
                'Product_Name': ['Laptop', 'Tablet', 'Monitor', 'Mouse', 'Keyboard', 'Phone'],
                'Category': ['Electronics', 'Electronics', 'Electronics', 'Accessories', 'Accessories', 'Electronics']}

product_df = pd.DataFrame(product_data)
```

```python
sales_df.head()
```

```python
product_df.tail()
```

```python
sales_df.info()
```

```python
product_df.info()
```

```python
# For the Units_Sold column, we’ll assume the missing value is 0 (no sales).
# For Sale_Amount, we’ll assume an average sale price

sales_df['Units_Sold'].fillna(0, inplace=True)
sales_df['Sale_Amount'].fillna(sales_df['Sale_Amount'].mean(), inplace=True)
sales_df
```

```python
sales_df['Price_Per_Unit'] = sales_df['Sale_Amount'] / sales_df['Units_Sold']
sales_df
```

```python
high_sales_df = sales_df[sales_df['Sale_Amount'] > 5000]
high_sales_df
```

```python
grouped_sales = sales_df.groupby(product_df['Category']).agg({
    'Sale_Amount': ['sum', 'mean'],
    'Units_Sold': ['sum']
    })

grouped_sales
```

```python
merged_df = pd.merge(sales_df, product_df, on='Product_ID')
merged_df
```

```python
pivot_table = merged_df.pivot_table(index='Category', values='Sale_Amount', aggfunc='sum')
pivot_table
```

```python
import matplotlib.pyplot as plt

merged_df.plot(x='Product_Name', y='Sale_Amount', kind='bar', title='Total Sales by Product')
plt.show()
```
