```python
import numpy as np
```

## Mean

```python
data = [15, 3, 12, 0, 24, 3]
mean = np.mean(data)

print(f"Mean: {mean}")
```

## Median

```python
data = [15, 3, 12, 0, 24, 3]
median = np.median(data)

print(f"Median: {median}")
```

## Mode

```python
from scipy import stats
data = [15, 3, 12, 0, 24, 3]
mode = stats.mode(data)
print(f"Mode: {mode.mode[0]}")
```

## Range

```python
data = [15, 3, 12, 1, 24, 3]
range_value = np.max(data) - np.min(data)
print(f"Range: {range_value}")
```

## Variance

```python
data = [15, 3, 12, 1, 24, 3]
variance = np.var(data)
print(f"Variance: {variance}")
```

## Standard Deviation

```python
data = [5, 7, 3, 8, 10]
std_dev = np.std(data)
print(f"Standard Deviation: {std_dev}")
```

## Quartiles

```python
quartiles = np.percentile(data, [3, 5, 7, 8, 10])
print(f"Quartiles: {quartiles}")
```

## Outliers and outlier detection

```python
z_scores = stats.zscore(data)
print(f"Z-Scores: {z_scores}")
```

```python
import matplotlib.pyplot as plt
plt.boxplot(data)
plt.title("Boxplot showing outliers")
plt.show()
```

## Skewness

```python
skewness = stats.skew(data)
print(f"Skewness: {skewness}")
```

## Covariance and Correlation

```python
data2 = [15, 18, 22, 25, 28]
covariance = np.cov(data, data2)[0, 1]
print(f"Covariance: {covariance}")
```

```python
correlation = np.corrcoef(data, data2)[0, 1]
print(f"Correlation: {correlation}")
```

## Example Using the Iris Dataset

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df.head()
```

## Mean

```python
# Calculate the mean of the Sepal Length.
mean_sepal_length = iris_df['sepal length (cm)'].mean()
print(f"Mean Sepal Length: {mean_sepal_length}")
```

### The mean sepal length is approximately 5.84 cm.

## Median

```python
# Calculate the median of Sepal Width.

median_sepal_width = iris_df['sepal width (cm)'].median()
print(f"Median Sepal Width: {median_sepal_width}")
```

### The median sepal width is 3.0 cm.

## Mode

```python
# Calculate the mode of Petal Width.

mode_petal_width = iris_df['petal width (cm)'].mode()[0]
print(f"Mode Petal Width: {mode_petal_width}")
```

### The mode of petal width is 0.2 cm.


## Range

```python
## Calculate the range of Sepal Length.

range_sepal_length = iris_df['sepal length (cm)'].max() - iris_df['sepal length (cm)'].min()
print(f"Range of Sepal Length: {range_sepal_length}")
```

### The range of sepal length is 3.6 cm.

## Variance

```python
## Calculate the variance of Petal Length.

variance_petal_length = iris_df['petal length (cm)'].var()
print(f"Variance of Petal Length: {variance_petal_length}")
```

## The variance of petal length is 3.11 cm².

```python
# Calculate the standard deviation of Sepal Width.

std_dev_sepal_width = iris_df['sepal width (cm)'].std()
print(f"Standard Deviation of Sepal Width: {std_dev_sepal_width}")
```

## Standard Deviation

```python
#  Calculate the standard deviation of Sepal Width.

std_dev_sepal_width = iris_df['sepal width (cm)'].std()
print(f"Standard Deviation of Sepal Width: {std_dev_sepal_width}")
```

## The standard deviation of sepal width is 0.43 cm.

## Quartiles

```python
# Calculate the first, second, and third quartiles for Petal Width.

q1, median, q3 = iris_df['petal width (cm)'].quantile([0.25, 0.5, 0.75])
print(f"Q1: {q1}, Median: {median}, Q3: {q3}")
```

##  Q1 = 0.3, Median = 1.3, Q3 = 1.8 cm.

## Z-score

```python
# Calculate Z-scores for Sepal Length.

from scipy import stats

z_scores_sepal_length = stats.zscore(iris_df['sepal length (cm)'])
print(z_scores_sepal_length)
```

### This will display the Z-scores, and any Z-score above 3 or below -3 indicates a potential outlier.

## Boxplot for Outlier Detection

```python
# Generate a boxplot for Sepal Length.

import matplotlib.pyplot as plt

plt.boxplot(iris_df['sepal length (cm)'])
plt.title('Boxplot of Sepal Length')
plt.show()
```

## Skewness

```python
# Calculate skewness of Sepal Length.

skewness_sepal_length = stats.skew(iris_df['sepal length (cm)'])
print(f"Skewness of Sepal Length: {skewness_sepal_length}")
```

```python
import seaborn as sns
sns.distplot(iris_df['sepal length (cm)'],hist=False)
```

### Since the value is close to 0, the distribution is nearly symmetric but with a slight positive skew.

## Covariance and Correlation

```python
# Calculate covariance between Sepal Length and Petal Length.

covariance = np.cov(iris_df['sepal length (cm)'], iris_df['petal length (cm)'])[0, 1]
print(f"Covariance: {covariance}")
```

### The covariance between Sepal Length and Petal Length in the Iris dataset is 1.24.
### This positive value suggests that as the Sepal Length increases, the Petal Length tends to increase as well.

```python
# Calculate correlation between Sepal Length and Petal Length.

correlation = iris_df['sepal length (cm)'].corr(iris_df['petal length (cm)'])
print(f"Correlation: {correlation}")
```

### The correlation between Sepal Length and Petal Length in the Iris dataset is 0.871.
### This value indicates a strong positive relationship.
### As Sepal Length increases, Petal Length also tends to increase in a proportionate manner.

#### - Central Tendency: We calculated the mean, median, and mode for Sepal and Petal measurements.
#### - Dispersion: We explored variance, standard deviation, and range for the dataset, and visualized it with quartiles.
#### - Outliers: Z-scores and boxplots helped us detect outliers.
#### - Skewness: Skewness indicates the distribution shape of Sepal Length.
#### - Relationships: Covariance and correlation between Sepal Length and Petal Length show a strong relationship.
