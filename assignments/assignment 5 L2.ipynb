#### Create the classes and functions as per following description: 
- Create a class Person : 
   - The attributes of the class are name, age, city
   - The functions of the class are :
       - __init__(self, name, age, city)
       - show_details(self)
       - update_city(self, new_city)
       
- Create a class Employee which inherits the class Person:
    - The attributes of the class are name , age, city, salary, position, dept
    - The functions of the class are : 
        - __init__(self, name , age, city, salary, position, dept)
        - show_details(self)
        - update_salary(self, new_salary)
        - update_dept(self, new_dept)
        - update_position(self, new_pos)
        - inc_salary(self)
        - dec_salary
        
- Make use of the super function when necessary
- Create objects and Demonstrate working of all the functions

```python
# creating a class called person with attributes name, age and city
# the functions of the class are __init__,show_details and update_city

class Person:                                      
    def __init__(self,name,age,city):
        self.name = name
        self.age = age
        self.city = city
        
    def show_details(self):
        print("The details are : ")
        print("Name     : ",self.name)
        print("Age      : ",self.age)
        print("City     : ",self.city)
        
    def update_city(self,new_city):
        self.city = new_city
        print('The city for {} has been changed to {}'.format(self.name, new_city))

# creating a child class Employee of the class Person
# the attributes of the child class are name , age, city, salary, position and dept
# the function of the class are __init__,show_details,update_salary, update_dept, update_position,inc_salary and dec_salary

class Employee(Person):
    def __init__(self,name,age,city,salary,position,dept):
        super().__init__(name,age,city)
        self.salary = salary
        self.position = position
        self.dept = dept
        
    def show_details(self):                        # for displaying the details of the employee
        super().show_details()
        print("Salary     : ",self.salary)
        print("Position   : ",self.position)
        print("Department : ",self.dept)
        print("\n")
        
    def update_salary(self, new_salary):          # for updating the salary
        self.salary = new_salary
        print('The salary for {} has been changed to     : {}'.format(self.name, new_salary))
        
    def update_position(self, new_pos):          # for updating position
        self.position = new_pos
        print('The position of {} has been changed to    : {}'.format(self.name, new_pos))
        
    def update_dept(self, new_dept):              # for updating department
        self.dept = new_dept
        print('The Department of {} has been changed to  : {}'.format(self.name, new_dept))
              
    def inc_salary(self):                        # for incrementing salary
        self.salary += 1000
        print('The salary for {} has been incremented to : {}'.format(self.name, self.salary))
        
    def dec_salary(self):                        # for decrementing salary
        self.salary -=1000
        print('The salary for {} has been decremented to : {}'.format(self.name, self.salary))

# creating the object of the class
emp=Employee("Abdul sajid",38,"Kasaragod",45000,"Teller","Finance")

# calling the functions of the class to print and update the details
emp.show_details()
emp.update_salary(50000)
emp.update_position("Cashier")
emp.update_dept("Accounts")
emp.inc_salary()
emp.dec_salary()
```
