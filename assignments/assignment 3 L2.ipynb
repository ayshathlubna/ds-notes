#### 1. Tom and Bob are playing a game of dice. Each time a multiple of two comes up Tom gets a point and if a multiple of 3 comes up bob gets a point. If a multiple of both comes up then a point is deducted from both. Write a program to take a list as input consisting of 10 outcomes of the dice throw. Print the scores of both Tom and bob and the winner.

```python
# a program to take a list as input consisting of 10 outcomes of the dice throw and find the winner

result=list(map(int,input("Enter the result : ").split()))

tom_score=0
bob_score=0

for i in result:
    if i%2==0 and i%3==0:           # If a multiple of 2 and 3 comes up then a point is deducted from both.
        tom_score -= 1
        bob_score -= 1
    elif i%2==0:                    # if a multiple of two comes up Tom gets a point 
        tom_score += 1
    elif i%3==0:                    # if a multiple of 3 comes up bob gets a point
        bob_score += 1
        
# Printing the score of both the players        

print("\nPoints Tom scored : ",tom_score)       
print("Points Bob scored : ",bob_score)     

# finding the winner from the total points received by each player

print("\n")

if tom_score>bob_score:
    print("Tom is the winner")
else:
    print("Bob is the winner")
```

#### 2. Write a program to take a list of 10 numbers as input in single line and find out the largest element.

```python
# a program to take a list of 10 numbers as input and find out the largest element.

num=list(map(int,input("Enter the numbers   : ").split()))
print("The list of numbers : ",num,end="\n\n")

largest=0
for i in num:
    if i>largest:
        largest=i

print("The largest number in the list : ",largest)
```

#### 3. Write a program to take a number as input and find out whether it is prime or not.

```python
# a program to find out whether a number is prime or not.

num = int(input('Enter the number : '))

if num < 2:      # since the prime number starts from 2 
    print(num,'is not a prime number')
else:
    for i in range(2, int(num ** (1/2)) + 1):
        if num % i == 0:
            print(num, 'is not a prime number')
            break
    else:
        print(num, 'is a prime number')
```

#### 4. Take a list and a number as input, print all consecutive pairs of two numbers from the list having sum equal to the given number.

```python
# print all consecutive pairs of two numbers from the list having sum equal to the specified number.

lists=list(map(int,input("Enter the numbers : ").split()))
n=int(input("Enter the number : "))

print("\nThe list of numbers : ",lists,end="\n\n")
print("The list of all consecutive pairs of 2 numbers having sum equal to the number {} : \n".format(n))

for i in range(0,len(lists)-1):
    if lists[i]+lists[i+1]==n:
        print( "({}, {}) ".format(lists[i],lists[i+1]),end=" ")

```

#### 5. Take a string as input and find out the number of words with even length and number of words with odd length.

```python
# A program to find out the number of words with even length and odd length in a string.

string=input("Enter the string : ").split()

print("\n")

even_count=0
odd_count=0

for i in string:
    if len(i)%2==0:
        even_count += 1
    else:
        odd_count += 1
        
print("The no.of words have even length : ",even_count)
print("The no.of words have odd length  : ",odd_count)
```

#### 7. Write a program to sort a given list using selection sort algorithm. (study selection sort and implement)

```python
# a program to sort a given list using selection sort algorithm.

def selectionsort(array,size):
    for i in range(size):
        min_index=i
        
        for j in range(i+1,size):
            if array[j]<array[min_index]:
                min_index=j
                
        array[i],array[min_index]=array[min_index],array[i]
        
lists=[8,0,6,1,2,-3,-11,9,2,-5]

n=len(lists)
print("The unsorted lists : ",lists)

selectionsort(lists,n)
print('The list after selection sort is : ',lists)

```

#### 8. Write a program to sort a given list using bubble sort algorithm. (Study bubble sort and implement)

```python
# a program to sort a given list using bubble sort algorithm

def bubblesort(lists):
    n=len(lists)
    
    swapped=False
    for i in range(n-1,0,-1):
        for j in range(i):
            if lists[j]>lists[j+1]:
                swapped=True
                lists[j],lists[j+1]=lists[j+1],lists[j]
        if not swapped:
            return
        
lists=[15,8,-17,0,12,-6,9,11,-1,4]

print("Unsorted list is : ",lists)

bubblesort(lists)
print("List after bubble sorting : ",lists)

```

#### 9. Take a string and a word and input. Find the frequenecy of occurrence of the word in the given string. (Without using any inbuilt function)

```python
# a program to find the frequenecy of occurrence of the word in the given string

string = input("Enter the string : ").split()
word = input("Enter the word : ")

count = 0

for w in string:
    if w == word:
        count += 1
        
print("\nThe number of times '{}' occured in the string : ".format(word),count)
```
