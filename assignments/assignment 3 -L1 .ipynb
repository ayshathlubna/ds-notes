### 1. 

Write a Python function that prints out the first n rows of Pascal's triangle.

- Input : 5
- Output: 
          1
         1 1
        1 2 1
       1 3 3 1
      1 4 6 4 1

```python
# a Python function that prints out the first n rows of Pascal's triangle.

def pascals(n):
    coef = 1

    for i in range(1, n+1):
        for space in range(1, n+1-i):
            print(" ",end="")
        for j in range(0, i):
            if j==0 or i==0:
                coef = 1
            else:
                coef = coef * (i - j)//j
            print(coef, end = " ")
        print()
        
pascals(5)
```

### 3. 

Write a program to print prime numbers from 1 to 100


```python
#  a program to print prime numbers from 1 to 100

print("The prime numbers between 1 to 100 are : ")
for num in range(2,100):
    
    for i in range(2,int(num**(1/2))+1):
        
        if num%i==0:
            break
    else:
            print(num,end=" ")      
```

### 4. 

Write a program to find given number is palindrome or not

Hint:
- 121 is palindrome
- 123 is not palindrome

```python
# a program to find given number is palindrome or not

num=input("Enter the number : ")
print(num)

for i in range(len(num)): # if the number and its reverse is same its palindrome
    if num[i]!=num[-1-i]:
        print(num," is not a palindrome")
        break
else:
    print(num," is palindrome")
        
```

Write a program to iterate the first 10 numbers and in each iteration, print the sum of the current and previous number.

- Current Number 0 Previous Number  0  Sum:  0
- Current Number 1 Previous Number  0  Sum:  1 and so on

```python
# a program to iterate the first 10 numbers and print the sum of the current and previous number

sum=0

for i in range(1,11):
    prev_num=i-1
    sum=prev_num+i
    print("The sum of current number {} and previous number {} : ".format(i,prev_num),sum)
    
```
