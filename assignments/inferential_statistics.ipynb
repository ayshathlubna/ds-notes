## One sample t-Test

```python
from scipy import stats

# Sample data (cookie weights)
cookie_weights = [29.8, 30.1, 30.2, 29.9, 30.0, 30.3]

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(cookie_weights, 30)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

## Result

- A larger t-statistic means the sample mean is farther from the population mean.Here the t-statistic is small (~0.65), indicating that the sample mean is close to the population mean of 30 grams.
- A p-value greater than 0.05 means that we fail to reject the null hypothesis. Here the p-value is ~0.54, which is quite high. This suggests that the difference between the sample mean and the population mean of 30 grams is not statistically significant.

## Two-Sample t-Test

```python
# Group A and Group B weight loss data
group_A = [2.1, 2.5, 2.8, 3.0, 3.2]
group_B = [1.8, 2.0, 2.1, 2.3, 2.5]

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(group_A, group_B)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

## Result
- The T-statistic of 2.54 indicates a significant difference in weight loss between Group A and Group B.
- The P-value of 0.0345 is less than the common significance level of 0.05. Therefore, we reject the null hypothesis, suggesting that there is statistically significant evidence that Group A experienced greater weight loss than Group B. This indicates that the method or treatment used in Group A is likely more effective in promoting weight loss compared to the method used in Group B.

## Paired t-Test

```python
# Before and after test scores
before = [70, 75, 72, 78, 74]
after = [78, 80, 82, 85, 84]

# Perform paired t-test
t_stat, p_value = stats.ttest_rel(before, after)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

## Result
- A large negative T-statistic -8.43 indicates a significant decrease in test scores before tutoring compared to after tutoring.
- The P-value 0.00108 is much smaller than 0.05, so we reject the null hypothesis.
- The extremely low P-value suggests that there is strong statistical evidence that the tutoring had a significant positive impact on the students' test scores. The negative T-statistic shows that the scores after tutoring are much higher than before tutoring. Therefore, the tutoring was effective in improving the students' performance.

## Chi-Square Test

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Contingency table: Gender vs Phone brand choice
data = pd.DataFrame({
    'Brand A': [30, 25],  # First row for Male, second row for Female
    'Brand B': [20, 35],
    'Brand C': [50, 40]
}, index=['Male', 'Female'])

# Perform Chi-Square test
chi2_stat, p_val, dof, ex = chi2_contingency(data)
print(f"Chi-Square Statistic: {chi2_stat}, P-value: {p_val}")
```

## Result

- Chi-Square Statistic 5.66 suggests there might be some association between gender and smartphone brand choice.
- The P-value 0.0591 is greater than the common significance level 0.05, so you fail to reject the null hypothesis. This means that, based on the data you have, the evidence isn't strong enough to claim a statistically significant association between gender and smartphone brand choice.

## ANOVA

```python
# Data: Test scores of students in 3 groups
group_A = [85, 88, 90, 93, 87]
group_B = [78, 82, 85, 80, 81]
group_C = [92, 94, 89, 95, 93]

# Perform ANOVA
f_stat, p_value = stats.f_oneway(group_A, group_B, group_C)
print(f"F-statistic: {f_stat}, P-value: {p_value}")
```

## Result
- The larger F-statistic 23.56 suggests that the differences between groups are large compared to the variability within groups.
- P-value: .0000699 is extremely small, much smaller than the typical significance level of 0.05.
- Since the P-value is much less than 0.05, we reject the null hypothesis. This means there is strong evidence to suggest that the teaching method significantly influences student test scores. At least one of the groups (A, B, or C) has a significantly different mean score compared to the others.

## one sample z test

```python
from statsmodels.stats.weightstats import ztest

# Sample data
sample_heights = [172, 171, 173, 170, 169, 172, 174, 171, 172, 170] * 5  # Total 50 data points

# Population parameters
population_mean = 170
population_std = 6  # known population std

# Perform Z-test
z_stat, p_value = ztest(sample_heights, value=population_mean, alternative='two-sided')

print(f"Z-statistic: {z_stat}, P-value: {p_value}")
```

## Result
-  Here the p-value is very small and Z-statistic is very high, so we can reject the null hypothesis. This indicates that the mean height of the sample is significantly different from the population mean of 170 cm.

## Two sample z test

```python
from statsmodels.stats.weightstats import ztest

# Sample data
class_A_heights = [172, 171, 173, 170, 169, 172, 174, 171, 172, 170] * 10  # n=100
class_B_heights = [174, 175, 173, 176, 174, 173, 175, 176, 174, 172] * 12  # n=120

# Perform two-sample Z-test
z_stat, p_value = ztest(class_A_heights, class_B_heights, alternative='two-sided')

print(f"Z-statistic: {z_stat}, P-value: {p_value}")
```

## Result
- Here the p-value is very small and the Z-statistic is large, we reject the null hypothesis. This means there is strong evidence that the average heights of Class A and Class B are significantly different.

## Hypothesis testing

```python
import numpy as np
from scipy import stats

# Test scores for both methods
traditional_scores = np.array([70, 75, 80, 68, 72, 78, 74, 71, 79, 73])
new_method_scores = np.array([85, 90, 88, 92, 86, 91, 89, 87, 93, 90])

# Perform the paired T-test
t_stat, p_value = stats.ttest_rel(traditional_scores, new_method_scores)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

## Result 

- The T-statistic −12.07 indicates that there is a significant difference between the scores from the traditional method and the new method, with the new method generally yielding higher scores.
- The P-value 0.0000007337 is extremely small, much less than the common significance level of 0.05.
-  Since the P-value is much less than 0.05, we reject the null hypothesis. This indicates that there is strong evidence that the new teaching method has a significantly positive impact on student test scores compared to the traditional method. In other words, students scored significantly higher with the new method.

```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Number of students
n_students = 200

# Generate data
student_ids = np.arange(1, n_students + 1)
ages = np.random.randint(18, 25, size=n_students)
genders = np.random.choice(['Male', 'Female'], size=n_students, p=[0.5, 0.5])
study_hours = np.random.normal(10, 3, size=n_students).round(1)  # mean=10, std=3
study_hours = np.clip(study_hours, 1, 20)  # Set limits on study hours
test_scores = np.random.normal(70, 15, size=n_students).round(1)  # mean=70, std=15
test_scores = np.clip(test_scores, 40, 100)  # Limit test scores between 40 and 100
tutoring = np.random.choice([0, 1], size=n_students, p=[0.7, 0.3])  # 30% students get tutoring
courses = np.random.choice(['Science', 'Arts', 'Commerce'], size=n_students, p=[0.4, 0.3, 0.3])
extracurricular = np.random.choice([0, 1], size=n_students, p=[0.6, 0.4])  # 40% participate in extracurricular activities
diet_type = np.random.choice(['Vegetarian', 'Non-Vegetarian', 'Vegan'], size=n_students, p=[0.3, 0.5, 0.2])
satisfaction = np.random.randint(1, 6, size=n_students)  # Satisfaction rating from 1 to 5

# Create DataFrame
data = pd.DataFrame({
    'Student_ID': student_ids,
    'Age': ages,
    'Gender': genders,
    'Study_Hours_per_Week': study_hours,
    'Test_Score': test_scores,
    'Tutoring': tutoring,
    'Course': courses,
    'Extracurricular': extracurricular,
    'Diet_Type': diet_type,
    'Satisfaction': satisfaction
})

# Show first 5 rows of the dataset

```
