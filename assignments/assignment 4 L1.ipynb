### 1. 

Create a function ``lets_find_88`` Given a list of ints, return True if the array contains a 8 next to a 8 somewhere.
- lets_find_88([1, 8, 8]) → True
- lets_find_88([1, 8, 1, 8]) → False
- lets_find_88([8, 1, 8]) → False

```python
# a program to create a function and return True if the array contains a 8 next to a 8 somewhere

# creating 3 lists
list1 = [1,8,8]
list2 = [1,8,1,8]
list3 = [8,1,8]

# checking whether the array has a 8 next to a 8 using function
def lets_find_88(x):
    for i in range(0,len(x)):
        if x[i]==8:
            if x[i+1]==8:
                return True
            else:
                return False
        
# printing the output by passing the arguments to the function
            
print("The list " + str(list1) + " has an 8 next to a 8 in the list : ",lets_find_88(list1))
print("The list " + str(list2) + " has an 8 next to a 8 in the list :",lets_find_88(list2))
print("The list " + str(list3) + " has an 8 next to a 8 in the list :",lets_find_88(list3))
```

### 3. 

Write a function ``how_many_primes`` that returns the number of prime numbers that exist up to and including a given number

- how_many_primes(100) --> 25

```python
# a program to count the number of prime numbers exist upto a given number

def how_many_primes(n):
    count = 0
    
    for num in range(2,n):
        for i in range(2,int(num**(1/2))+1):
            if num % i==0:     # checking whether the number is prime or not
                 break
        else:
            count += 1  # incrementing the value of count everytime a prime number is found

    return count # will return the total number of prime numbers when the function is being called
        

print("The number of primes exist",how_many_primes(100))
```
