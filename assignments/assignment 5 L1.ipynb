### 1. 

Write a Program to create a class 'Company' and using the class instance and print all the writable attributes of that object such as age, gender, salary, profession and designation. We take atleast 5 employees of an organization

```python
# a program to create a class 'Company' to store details of 5 employees of an organization

# Creating the class Company with attributes name, age, gender, salary, profession and designation
# the functions of the class are __init__ and show_details

class Company:               
    def __init__(emp, name, age, gender, salary, profession, designation):
        emp.name = name
        emp.age = age
        emp.gender = gender
        emp.salary=salary
        emp.profession=profession
        emp.designation=designation
        
    def show_details(emp):    # this function will display all the details of the employee
        print("\nThe details of the employee {} is : ".format(emp.name))
        print("Name        : ",emp.name)
        print("Age         : ",emp.age)
        print("Gender      : ",emp.gender)
        print("Salary      : ",emp.salary)
        print("Profession  : ",emp.profession)
        print("Designation : ",emp.designation)

# creating the objects of the class and passing the values to the attributes

emp1 = Company("Musthafa",45,"Male",55000,"Manager","Zone Manager")
emp2 = Company("Sajid",39 ,"Male",46000 ," Manager", "Area Manager ")
emp3 = Company("Abdulla",31 ,"Male",35000 ,"Manager", "Branch Manager")
emp4 = Company("Laila",42 ,"Female",30000,"Manager", "Assistant Branch Manager")
emp5 = Company("Mufeeda",34,"Female",40000 ,"Manager", "Assistant Area Manager")

# calling the show_details() function of the class

emp1.show_details()
emp2.show_details()
emp3.show_details()
emp4.show_details()
emp5.show_details()
```

### 3. 

Build a Simple Calculator using Exception Handling (try, except) that solves 4 arithmetic operations like Addition, Subtraction, Multiplication and Division.


```python
# a program to build a Simple Calculator using Exception Handling (try, except) that solves 4 arithmetic operations like Addition, Subtraction, Multiplication and Division.

# creating a class named Calculator
# the functions of the class are __init__,addition,subtraction,multiplication and division

class Calculator():
    def __init__(self,a,b):                   # this function will take the value for two variables a and b
        self.a = a
        self.b = b
        
    def addition(self):                       # function for performing addition
        add = self.a + self.b
        return add
    
    def subtraction(self):                    # function for performing subtraction
        sub = self.a - self.b
        return sub
    
    def multiplication(self):                 # function for performing multiplication
        mult = self.a * self.b
        return mult
    
    def division(self):                       # function for performing division and using exception handling to handle zero division error
        try:
            div = self.a / self.b
            return div
        except ZeroDivisionError:
            print("Division by zero not possible")

# taking values to the variables a and b
a = int(input("Enter the first number  : "))
b = int(input("Enter the second number : "))

x=Calculator(a,b)                           # creating the object of the class

op=1
while op!=0:                                # coding to perform the arithmetic operations based on the conditions
    op = input("Enter the operator (+,-,*,/) or zero to exit: ")
    if op == "+":
        print("Sum        : ",x.addition())
    elif op == "-":
        print("Difference : ",x.subtraction())
    elif op == "*":
        print("Product    : ",x.multiplication())
    elif op == "/":
        print("Quotient    : ",x.division()) 
    elif op=="0":
        print("Exit")
        break
    else:
        print("Invalid choice")

```
