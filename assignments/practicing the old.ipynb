- lets_find_88([1, 8, 8]) → True
- lets_find_88([1, 8, 1, 8]) → False
.-lets_find_88([8, 1, 8]) → False

```python
def lets_find_88(list1):
    for i in range(len(list1)-1): 
        if list1[i]==8 and list1[i+1]==8:
            return True
    return False      
```

## map(),filter() and zip()

```python
def square(x):
    return x ** 2

numbers = [1, 2, 3, 4, 5]

squared = list(map(square, numbers))

print(squared) 

# The square function is defined to return the square of a number.
# map() applies the square function to every element in the list numbers
```

```python
def is_even(x):
    return x % 2 == 0

numbers = [1, 2, 3, 4, 5]

even_numbers = list(filter(is_even, numbers))

print(even_numbers)  

# The is_even function checks if a number is even (x % 2 == 0).
# filter() keeps only the elements for which is_even returns True.
```

```python
names = ['Alice', 'Bob', 'Charlie']
scores = [85, 92, 78]

paired = list(zip(names, scores))

print(paired) 

# zip() pairs the first elements of names and scores, then the second elements, and so on.
# The result is a list of tuples, each containing a name and a corresponding score.
```

```python
def square(x):
    return x ** 2

def is_even(x):
    return x % 2 == 0

numbers = [1, 2, 3, 4, 5]

# Filter even numbers first
even_numbers = list(filter(is_even, numbers))

# Apply map to square the even numbers
even_squared = list(map(square, even_numbers))

# Pair the original even numbers with their squares
paired = list(zip(even_numbers, even_squared))

print(paired)

# First: The filter() function filters the original list numbers to get only the even numbers.
# Next: The map() function squares these filtered even numbers.
# Finally: The zip() function pairs each original even number with its corresponding square.
```

## using map() and filter() to get even numbers

```python
def check_even(num):
    if num % 2 == 0:
        return True
    else:
        return False
```

```python
nums = [11,33,22,44,66,55,77,88]
```

```python
list(map(check_even, nums))
```

```python
list(filter(check_even, nums))
```

```python
# The map() function returns a list of boolean values indicating whether each number in nums is even or not.
# The filter() function returns a list of only the even numbers from nums.
```

## Check for prime numbers

```python
def check_prime(num):
    if num < 2:
        return False
    else:
        for i in range(2, int(num ** (1/2)) + 1):
            if num % i == 0:
                return False
        else:
            return True    
```

```python
list(filter(check_prime,range(0,100)))
```

```python
# The check_prime function determines if a number is prime by checking divisibility up to the square root of the number.
# The filter() function uses check_prime to find all prime numbers within the specified range (0 to 99),
# and returns a list of these primes.
```

```python
check_prime(2)
```

# Zip()

```python
icecreams = ['Butterscotch', 'Chocolate', 'Vanilla']
nums = [5, 4, 3]
prices = [80,50,40]

list(zip(icecreams, nums, prices))
```

```python
''' 
- The zip() function is used to combine the three lists into a single iterable, where each element is a tuple containing 
  elements from the corresponding positions in each list.

- Specifically, the zip() function will take the first element from each list and group them together in a tuple,
  then the second element from each list, and so on.

- Result of zip():
        The zip() function will combine the three lists as follows:
        The first tuple will be: ('Butterscotch', 5, 80), which combines the first elements of icecreams, nums, and prices.
        The second tuple will be: ('Chocolate', 4, 50), which combines the second elements of icecreams, nums, and prices.
        The third tuple will be: ('Vanilla', 3, 40), which combines the third elements of icecreams, nums, and prices.
        
        The zip() function creates an iterable, so you convert it to a list using list().
'''
```

## Lambda expression and List comprehension

## Squared Numbers

```python
# Using Lambda with map:

numbers = [1, 2, 3, 4]
doubled_lambda = list(map(lambda x: x * 2, numbers))
print("using lambda with map : ",doubled_lambda) 

# Using List Comprehension:

doubled_list_comp = [x * 2 for x in numbers]
print("Using list comprehension : ",doubled_list_comp)  
```

## Filtering Even Numbers

```python
# Using Lambda with filter:

numbers = [1, 2, 3, 4, 5, 6]
evens_lambda = list(filter(lambda x: x % 2 == 0, numbers))
print("using lambda with filter : ",evens_lambda)  

# Using List Comprehension:

evens_list_comp = [x for x in numbers if x % 2 == 0]
print("Using list comprehension : ",evens_list_comp)  
```

## Combining Two Lists Element-wise

```python
# Using lambda with map
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined_lambda = list(map(lambda x, y: x + y, list1, list2))


# Using list comprehension
combined_list_comp = [x + y for x, y in zip(list1, list2)]

print("combining 2 lists using lambda with map : ",combined_lambda)
print("combining 2 lists using list comprehension : ",combined_list_comp)  
```

## Creating a Dictionary from Two Lists

```python
# Using lambda with map and zip
keys = ['a', 'b', 'c']
values = [1, 2, 3]
dict_lambda = dict(map(lambda k, v: (k, v), keys, values))


# Using list comprehension
dict_comp = {k: v for k, v in zip(keys, values)}

print("creating a dictionary from  2 lists using lambda with map and zip : ",dict_lambda)
print("creating a dictionary from  2 lists using list comprehension : ",dict_comp)  
```

## Applying a Function to Each Element in a List

```python
def add_five(x):
    return x + 5

# Using lambda with map

numbers = [10, 20, 30]
add_five_lambda = list(map(lambda x: add_five(x), numbers))


# Using list comprehension
add_five_list_comp = [add_five(x) for x in numbers]

print("Applying a Function to Each Element in a List using lambda with map : ",add_five_lambda)
print("Applying a Function to Each Element in a List using list comprehension : ",add_five_list_comp)  
```

##  Generating a List of Tuples (Index, Value)

```python
names = ['Alice', 'Bob', 'Charlie']

# Using lambda with enumerate

indexed_names_lambda = list(map(lambda x: (x[0], x[1]), enumerate(names)))

# Using list comprehension
indexed_names_list_comp = [(index, name) for index, name in enumerate(names)]

print("Generating a List of Tuples using lambda with enumerate : ",indexed_names_lambda)
print("Generating a List of Tuples using list comprehension : ",indexed_names_list_comp)  
```

## Converting Strings to Uppercase

```python
strings = ['apple', 'banana', 'cherry']

# Using lambda with map

upper_lambda = list(map(lambda x: x.upper(), strings))

# Using list comprehension
upper_list_comp = [x.upper() for x in strings]

print("Converting Strings to Uppercase using lambda with map : ",upper_lambda)
print("Converting Strings to Uppercase using list comprehension : ",upper_list_comp)  
```

## Filtering and Modifying Strings

```python
strings = ['apple', 'banana', 'cherry']

# Using lambda with map and filter

filtered_upper_lambda = list(map(lambda x: x.upper(), filter(lambda x: 'a' in x, strings)))

# Using list comprehension
filtered_upper_list_comp = [x.upper() for x in strings if 'a' in x]

print("Filtering and Modifying Strings using lambda with map : ",filtered_upper_lambda)
print("Filtering and Modifying Strings using list comprehension : ",filtered_upper_list_comp)  
```
