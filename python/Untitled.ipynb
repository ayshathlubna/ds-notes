```python
# Importing necessary module
import os

# Step 1: Define the file path
file_path = 'example_file.txt'

# Step 2: Check if the file exists before attempting to open it
try:
    # Try to open the file in read mode
    with open(file_path, 'r') as file:
        # Step 3: Attempt to read the contents of the file
        content = file.read()
        print("File content successfully read.")
        
        # Simulate a processing step that might raise an error (e.g., ValueError)
        if content == "":
            raise ValueError("The file is empty.")
        
        print("File Content:\n", content)
        
except FileNotFoundError:
    # Step 4: Handle the case where the file does not exist
    print(f"Error: The file '{file_path}' does not exist.")
    
except PermissionError:
    # Step 5: Handle the case where the file cannot be accessed due to permission issues
    print(f"Error: You do not have permission to read the file '{file_path}'.")
    
except ValueError as ve:
    # Step 6: Handle any value errors that might occur during file processing
    print(f"Error: {ve}")
    
except Exception as e:
    # Step 7: Catch-all for any other exceptions that are not explicitly handled
    print(f"An unexpected error occurred: {e}")
    
else:
    # Step 8: This block will execute if no exceptions are raised
    print("File processed successfully without any errors.")
    
finally:
    # Step 9: This block will execute no matter what, ensuring that any necessary cleanup is done
    print("Execution completed. Whether an error occurred or not, this block is always executed.")

```

```python
# Step 10: Attempt to create and write to the file if it doesn't exist
try:
    # Try to write to the file in write mode (this will create the file if it doesn't exist)
    with open(file_path, 'w') as file:
        file.write("This is an example of file handling in Python.\n")
        file.write("Understanding how to handle exceptions is crucial for writing robust code.\n")
    print("File written successfully.")
    
except PermissionError:
    # Step 11: Handle the case where the file cannot be written due to permission issues
    print(f"Error: You do not have permission to write to the file '{file_path}'.")
    
except Exception as e:
    # Step 12: Catch-all for any other exceptions during writing
    print(f"An unexpected error occurred while writing to the file: {e}")
    
finally:
    # Step 13: Cleanup, confirming that the file operation is complete
    print("File operation completed.")

# Step 14: Re-read the file to confirm the content written
try:
    with open(file_path, 'r') as file:
        content = file.read()
        print("Re-reading the file after writing:")
        print("File Content:\n", content)
except FileNotFoundError:
    print(f"Error: The file '{file_path}' does not exist.")
except PermissionError:
    print(f"Error: You do not have permission to read the file '{file_path}'.")
except Exception as e:
    print(f"An unexpected error occurred while reading the file: {e}")
```
