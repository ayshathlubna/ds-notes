## Method 1: Using a Temporary Variable

```python
# Take two numbers as input and swap their values using a temporary variable
a = int(input("Enter the value of a: "))  # Input value for a
b = int(input("Enter the value of b: "))  # Input value for b

print("Value of a before swap: ", a)  # Print initial value of a
print("Value of b before swap: ", b)  # Print initial value of b

# Swap the values using a temporary variable
temp = a  # Store the value of a in temp
a = b     # Assign the value of b to a
b = temp  # Assign the value of temp (initial value of a) to b

print("Value of a after swap: ", a)  # Print value of a after swap
print("Value of b after swap: ", b)  # Print value of b after swap
```

## Method 2: Using Tuple Unpacking

```python
# Take two numbers as input and swap their values using tuple unpacking
a = int(input("Enter the value of a: "))  # Input value for a
b = int(input("Enter the value of b: "))  # Input value for b

print("Value of a before swap: ", a)  # Print initial value of a
print("Value of b before swap: ", b)  # Print initial value of b

# Swap the values using tuple unpacking
a, b = b, a  # Assign the value of b to a and the value of a to b

print("Value of a after swap: ", a)  # Print value of a after swap
print("Value of b after swap: ", b)  # Print value of b after swap
```

## Method 3: Using Arithmetic Operations

```python
# Take two numbers as input and swap their values using arithmetic operations
a = int(input("Enter the value of a: "))  # Input value for a
b = int(input("Enter the value of b: "))  # Input value for b

print("Value of a before swap: ", a)  # Print initial value of a
print("Value of b before swap: ", b)  # Print initial value of b

# Swap the values using arithmetic operations
a = a + b  # a holds the sum of a and b
b = a - b  # Subtracting b from sum of a and b gives the initial value of a
a = a - b  # Subtracting the new value of b (initial value of a) from the sum gives the initial value of b

print("Value of a after swap: ", a)  # Print value of a after swap
print("Value of b after swap: ", b)  # Print value of b after swap
```
