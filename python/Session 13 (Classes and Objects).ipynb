```python
class Person:
    def __init__(self):
        self.first_name = None
        self.last_name  = None
        self.age = None
        self.city = None
    
    def set_details(self, fname, lname, age, city):
        self.first_name = fname
        self.last_name  = lname
        self.age = age
        self.city = city
        
    def show_details(self):
        print('\nThe details are : ')
        print('First Name : ', self.first_name)
        print('Last Name  : ', self.last_name)
        print('Age        : ', self.age)
        print('City       : ', self.city)
        
    def update_city(self, new_city):
        self.city = new_city
        print('The city for {} has been changed to {}'.format(self.first_name + ' ' + self.last_name, new_city))
```

```python
p1 = Person()
```

```python
p1.show_details()
```

```python
p1.set_details('Smeet', 'Daftary', 26, 'Delhi')
```

```python
p1.show_details()
```

```python
p1.update_city('Hyderabad')
```

```python
p1.show_details()
```

```python
p2 = Person()
p2.set_details('Jagjiban', 'Swan', 36, 'Patna')
p2.show_details()
```

```python
class Employee(Person):
    pass
```

```python
e1 = Employee()
```

```python
e1.show_details()
```

```python
e1.set_details('Ankit', 'Das', 35, 'Delhi')
e1.show_details()
```
