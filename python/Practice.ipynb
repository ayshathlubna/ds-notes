```python
def pattern(n):
    for i in range(1,n+1):
        for j in range(2*(n-i)):
            print(' ', end = '')
        for k in range(2*i - 1):
            print('*', end = '')
        print()
        
pattern(5)
```

```python
# Print highest and lowest scores
d1 = {'tom':[95,70,60,85,90],'bob':[50,60,70,80,90],'jon':[45,55,65,75,85]}
tom = sum(d1['tom'])
bob = sum(d1['bob'])
jon = sum(d1['jon'])

if tom > bob and tom > jon:
    print('Highest scorer is Tom with marks : ', tom)
elif bob > jon:
    print('Highest scorer is Bob with marks : ', bob)
else:
    print('Highest scorer is Jon with marks : ', jon)
    
if tom < bob and tom < jon:
    print('Lowest scorer is Tom with marks : ', tom)
elif bob < jon:
    print('Lowest scorer is Bob with marks : ', bob)
else:
    print('Lowest scorer is Jon with marks : ', jon)
```

```python
def total(nums):
    result = 0
    for i in nums:
        result += i
    return result
```
