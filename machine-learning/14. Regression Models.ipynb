## -------------------------------------------------------------------------------------------------

```python
cols = ['Vehicle Class', 'Transmission', 'Fuel Type', 'Cylinders', 'Engine Size(L)','Fuel Consumption Comb (L/100 km)']
X = data[cols]
Y = data[['CO2 Emissions(g/km)']]
```

```python
X.head(10)
```

```python
Y.head(10)
```

```python
print('Independent Feature Set Shape : ', X.shape)
print('Dependent Feature Set Shape   : ', Y.shape)
```

```python
# Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols = ['Engine Size(L)','Fuel Consumption Comb (L/100 km)']
X[cols] = scaler.fit_transform(X[cols])
X.head(10)
```

```python
# Dealing with categorical values
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
cols = ['Vehicle Class', 'Transmission', 'Fuel Type']
X[cols] = encoder.fit_transform(X[cols])
X.head(10)
```

```python
# Splitting the data into train and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
print('Training Data Shape   : ', x_train.shape)
print('Training Labels Shape : ', y_train.shape)
print('Testing  Data Shape   : ', x_test.shape)
print('Testing Labels Shape  : ', y_test.shape)
```

# ---------------------------------------------------------------------------------
--------------
# LinearRegression (OLS)

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
print('Training set score : ', regressor.score(x_train, y_train))
print('Testing set score  : ', regressor.score(x_test, y_test))
```

# ---------------------------------------------------------------------------------
-------------------
# SGDRegressor (Gradient Descent)

```python
from sklearn.linear_model import SGDRegressor
regressor = SGDRegressor(alpha = 0.001)
regressor.fit(x_train, y_train)
```

```python
print('Training set score : ', regressor.score(x_train, y_train))
print('Testing set score  : ', regressor.score(x_test, y_test))
```

# ---------------------------------------------------------------------------------
--------------------
# Regularized Regression Models
- 1. Lasso (L1 Regularization)
- 2. Ridge (L2 Regularization)

<img align = 'left' src = 'regularization.png' style = 'width:400px; hieght:150px' />

```python
from sklearn.linear_model import Lasso
regressor = Lasso(alpha = 1)
regressor.fit(x_train, y_train)

print('Training set score : ', regressor.score(x_train, y_train))
print('Testing set score  : ', regressor.score(x_test, y_test))
```

```python
from sklearn.linear_model import Ridge
regressor = Ridge(alpha = 1)
regressor.fit(x_train, y_train)

print('Training set score : ', regressor.score(x_train, y_train))
print('Testing set score  : ', regressor.score(x_test, y_test))
```

# ---------------------------------------------------------------------------------
----------------
# KNN Regression

```python
from sklearn.neighbors import KNeighborsRegressor
regressor = KNeighborsRegressor(n_neighbors = 3)
regressor.fit(x_train, y_train)

print('Training set score : ', regressor.score(x_train, y_train))
print('Testing set score  : ', regressor.score(x_test, y_test))
```

# ----------------------------------------------------------------------------------
---------------
# DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(max_depth = 6)
regressor.fit(x_train, y_train)

print('Training set score : ', regressor.score(x_train, y_train))
print('Testing set score  : ', regressor.score(x_test, y_test))
```

# --------------------------------------------------------------------------------
-----------------
# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 50, max_depth = 5)
regressor.fit(x_train, y_train)

print('Training set score : ', regressor.score(x_train, y_train))
print('Testing set score  : ', regressor.score(x_test, y_test))
```

# -----------------------------------------------------------------------------------
# LGBMRegressor

```python
from lightgbm import LGBMRegressor
regressor = LGBMRegressor(n_estimators = 50)
regressor.fit(x_train, y_train)

print('Training set score : ', regressor.score(x_train, y_train))
print('Testing set score  : ', regressor.score(x_test, y_test))
```
