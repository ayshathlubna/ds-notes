```python
cols = ['Vehicle Class', 'Transmission', 'Fuel Type', 'Cylinders', 'Engine Size(L)','Fuel Consumption Comb (L/100 km)']
X = data[cols]
Y = data[['CO2 Emissions(g/km)']]
```

```python
X.head(10)
```

```python
Y.head(10)
```

```python
print('Independent Feature Set Shape : ', X.shape)
print('Dependent Feature Set Shape   : ', Y.shape)
```

```python
# Splitting the data into train and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
print('Training Data Shape   : ', x_train.shape)
print('Training Labels Shape : ', y_train.shape)
print('Testing  Data Shape   : ', x_test.shape)
print('Testing Labels Shape  : ', y_test.shape)
```

```python
# Scaling 
cols = ['Engine Size(L)', 'Fuel Consumption Comb (L/100 km)']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Fitting on the training set
scaler.fit(x_train[cols])
# Transforming the training and testing
x_train[cols] = scaler.transform(x_train[cols])
x_test[cols]  = scaler.transform(x_test[cols])
```

```python
# Encoding
cols = ['Vehicle Class', 'Fuel Type', 'Transmission']
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
# Fit transform on train set
x_train[cols] = encoder.fit_transform(x_train[cols])
# Transform on test set
x_test[cols] = encoder.transform(x_test[cols])
```

# ================================================
# Hyperparameter Tuning

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
```

```python
search_params = {
    'n_estimators' : [10, 20, 30, 40, 50],
    'max_depth'    : [3, 4, 6, 8, 10],
    'max_features' : [4, 5, 6]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator = regressor,
                           param_grid = search_params,
                           scoring = ['r2'],
                           cv = 5,
                           refit = 'r2',
                           verbose = 3)

grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
grid_search.cv_results_
```

```python
df = pd.DataFrame(grid_search.cv_results_).sort_values('mean_test_r2', ascending = False)
df[20:40]
```

```python
df.to_csv('GridSearchResults.csv')
```

```python
# Best model after considering efficiency
best_model = RandomForestRegressor(max_depth = 6, n_estimators = 10, max_features = 5)
best_model.fit(x_train, y_train)

print('Training Set Score : ', best_model.score(x_train, y_train))
print('Testing Set Score  : ', best_model.score(x_test, y_test))
```

# RandomizedSearchCV

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
```

```python
search_params = {
    'n_estimators' : [10, 20, 30, 40, 50],
    'max_depth'    : [3, 4, 6, 8, 10],
    'max_features' : [4, 5, 6]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
randomized_search = RandomizedSearchCV(estimator = regressor,
                                       param_distributions = search_params,
                                       scoring = ['r2'],
                                       cv = 5,
                                       refit = 'r2',
                                       n_iter = 15,
                                       verbose = 3)

randomized_search.fit(x_train, y_train)
```

```python
randomized_search.best_params_
```

```python
randomized_search.best_score_
```

```python
best_model = randomized_search.best_estimator_
best_model
```

```python
df2 = pd.DataFrame(randomized_search.cv_results_).sort_values('mean_test_r2', ascending = False)
df2.head()
```

```python
df2.to_csv('RandomizedSearchResults.csv')
```

```python
# Considering efficiency
# Best model after considering efficiency
best_model = RandomForestRegressor(max_depth = 6, n_estimators = 10, max_features = 5)
best_model.fit(x_train, y_train)

print('Training Set Score : ', best_model.score(x_train, y_train))
print('Testing Set Score  : ', best_model.score(x_test, y_test))
```
