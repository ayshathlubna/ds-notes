```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
data = pd.read_csv('adult.csv')
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data.gender.value_counts()
```

```python
data.occupation.value_counts()
```

```python
data.groupby('occupation')['gender'].value_counts()
```

```python
pd.crosstab(data.occupation, data.gender, margins = True)
```

```python
observed = pd.crosstab(data.occupation, data.gender)
```

```python
# chi2 test_statistic, p-value, degrees of freedom, expected_frequences
result = chi2_contingency(observed)
result
```

```python
chi2_test_stat = result[0]
deg_freedom = result[2]
expected_freq = result[3]
```

```python
print(2809*16192 / 48842)
print(5611*16192 / 48842)
print(2809*32650 / 48842)
print(5611*32650 / 48842)
```

```python
alpha = 0.1
chi2_critical = chi2.ppf(alpha, deg_freedom)
chi2_critical
```

```python
if chi2_test_stat > chi2_critical:
    print('Reject Null Hypothesis')
else:
    print('Failed to reject Null Hypothesis')
```

## Education and Gender

```python
data.education.unique()
```

```python
pd.crosstab(data.education, data.gender, margins = True)
```

```python
observed = pd.crosstab(data.education, data.gender)
```

```python
# chi2 test_statistic, p-value, degrees of freedom, expected_frequences
result = chi2_contingency(observed)
chi2_test_stat = result[0]
deg_freedom = result[2]
expected_freq = result[3]
result
```

```python
alpha = 0.1
chi2_critical = chi2.ppf(alpha, deg_freedom)
chi2_critical
```

```python
if chi2_test_stat > chi2_critical:
    print('Reject Null Hypothesis')
else:
    print('Failed to reject Null Hypothesis')
```

### Effect of significance level and degree of freedom

```python
for i in [0.05, 0.1, 0.5, 0.99]:
    for j in range(1,16,5):
        print('For Significance Level : {} and Degree of Freedom : {} Chi2 Critical is = {}'.format(i,j,chi2.ppf(i,j)))
```
