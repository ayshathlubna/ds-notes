```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Calculating Z score
def z_score(sample_size, sample_mean, pop_mean, pop_std):
    std_error = pop_std / sample_size ** 0.5
    z = (sample_mean - pop_mean) / std_error
    return z
```

**Q-1:** A Shop selling Hyderabadi biryani claims that each parcel packet has 500 grams of biryani. You are sceptic of their claims and believe that on an average each packet does not contain 500 grams of biryani. How do you prove your claim? Given the population standard deviation is 50.

```python
l = [490, 220, 470, 500, 495, 496, 496, 498, 508, 480]
pop_mean = 500
pop_std = 50
sample_size = len(l)
sample_mean = sum(l) / sample_size
print('Sample Size : ', sample_size)
print('Sample Mean : ', sample_mean)
```

```python
# Calculate z score
z = z_score(sample_size, sample_mean, pop_mean, pop_std)
z
```

```python
# Lower tailed test
# calculating p value
p_value = norm.cdf(z)
p_value
```

```python
# Conclusion
alpha = 0.05
if p_value < alpha:
    print('We have sufficient evidence to reject null hypothesis.')
    print("Therefore the shop's claim to have 500 grams biryani per packet is false")
else:
    print('We fail to reject null hypothesis')
```
