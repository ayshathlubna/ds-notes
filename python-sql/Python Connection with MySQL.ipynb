```python
def insert_data():
    print('\nEnter the details of the employee to be addaed : ')
    empid = int(input('Enter the employee id : '))
    name = input('Enter the name : ')
    age = int(input('Enter the age : '))
    city = input('Enter the city : ')
    
    testdb_connector = mysql.connector.connect(
                        host = 'localhost',
                        user = 'root',
                        password = 'root',
                        database = 'testdb')

    mycursor = testdb_connector.cursor()
    
    query = '''INSERT INTO employee (empid, name, age, city)
           VALUES (%s, %s, %s, %s)'''
    val = (empid, name, age,city)
    mycursor.execute(query, val)
    
    testdb_connector.commit()
    testdb_connector.close()
    print('\n Row inserted Successfully')
```

```python
def show_data():
    testdb_connector = mysql.connector.connect(
                        host = 'localhost',
                        user = 'root',
                        password = 'root',
                        database = 'testdb')

    mycursor = testdb_connector.cursor()
    
    query = 'SELECT * FROM employee'
    mycursor.execute(query)
    
    print('\n')
    for row in mycursor:
        print(row)
    
    testdb_connector.commit()
    testdb_connector.close()
```

```python
def update_data():
    print('\nEnter the update details : ')
    print()
    column = input('Enter the column : ')
    value  = input('Enter the value : ')
    empid  = int(input('Enter the employee id : '))
    
    testdb_connector = mysql.connector.connect(
                        host = 'localhost',
                        user = 'root',
                        password = 'root',
                        database = 'testdb')

    mycursor = testdb_connector.cursor()
    
    if column == 'age':
        query = "UPDATE employee SET {} = {} WHERE empid = {}".format(column, int(value), empid)
    else:
        query = "UPDATE employee SET {} = '{}' WHERE empid = {}".format(column, value, empid)
        
    mycursor.execute(query)
    testdb_connector.commit()
    testdb_connector.close()
    print('\nUpdate Successfull')
```

```python
def delete_data():
    empid = int(input('Enter the employee id: '))
    
    testdb_connector = mysql.connector.connect(
            host = 'localhost',
            user = 'root',
            password = 'root',
            database = 'testdb3'
    )

    mycursor = testdb_connector.cursor()
    
    query = "DELETE FROM employee WHERE empid = {}".format(empid)
    mycursor.execute(query)
    
    testdb_connector.commit()
    testdb_connector.close()
    
    print('\nDeletion Successfull')
```
